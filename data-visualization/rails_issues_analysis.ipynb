```python
# libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv

def plotHorizontalLillipop(x, y, xlabel, ylabel):
  my_range=range(1,len(x)+1)
 
  # The horizontal plot is made using the hline function
  plt.hlines(y=my_range, xmin=0, xmax=y, color='skyblue')
  plt.plot(y, my_range, "o")
 
  # Add titles and axis names
  plt.yticks(my_range, x)
  plt.xlabel(xlabel)
  plt.ylabel(ylabel)

  # Show the plot
  plt.show()

def calculateIssuesPerYear():
  years=[]
  number_of_issues=[]

  with open('issues_per_year.csv','r') as csvfile:
      lines = csv.reader(csvfile, delimiter=',')
      for (row) in lines:
          years.append(row[0])
          number_of_issues.append(int(row[1]))

  plotHorizontalLillipop(x=years, y=number_of_issues, xlabel='# of issues', ylabel='Years')

def calculateIssuesPerReporter():
  reporters=[]
  number_of_issues=[]

  with open('issues_per_reporter.csv','r') as csvfile:
      lines = csv.reader(csvfile, delimiter=',')
      for (row) in lines:
          reporters.append(row[0])
          number_of_issues.append(int(row[1]))
      
  plotHorizontalLillipop(x=reporters, y=number_of_issues, xlabel='# of issues', ylabel='Reporters')

def calculateIssuesPerCategory():
  categories=[]
  number_of_issues=[]

  with open('issues_per_category.csv','r') as csvfile:
      lines = csv.reader(csvfile, delimiter=',')
      for (row) in lines:
          categories.append(row[0])
          number_of_issues.append(int(row[1]))
      
  plotHorizontalLillipop(x=categories, y=number_of_issues, xlabel='# of issues', ylabel='Categories')

calculateIssuesPerYear()
calculateIssuesPerReporter()
calculateIssuesPerCategory()
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
 

df2017 = pd.read_csv("issues_per_year_2017.csv")
df2018 = pd.read_csv("issues_per_year_2018.csv")
df2019 = pd.read_csv("issues_per_year_2019.csv")
df2020 = pd.read_csv("issues_per_year_2020.csv")
df2021 = pd.read_csv("issues_per_year_2021.csv")
df2022 = pd.read_csv("issues_per_year_2022.csv")
x2017 = df2017.number_of_issues.to_numpy();
x2018 = df2018.number_of_issues.to_numpy();
x2019 = df2019.number_of_issues.to_numpy();
x2020 = df2020.number_of_issues.to_numpy();
x2021 = df2021.number_of_issues.to_numpy();
x2022 = df2022.number_of_issues.to_numpy();

labels = df2017.month.to_numpy()

# set width of bars
barWidth = 0.15

# Set the size of the chart
plt.figure(figsize=(12, 6), dpi=80)
 
# Set position of bar on X axis
r1 = np.arange(len(x2017))
r2 = [x + barWidth for x in r1]
r3 = [x + barWidth for x in r2]
r4 = [x + barWidth for x in r3]
r5 = [x + barWidth for x in r4]
r6 = [x + barWidth for x in r5]
 
# Make the plot
plt.bar(r1, x2017, color='blue', width=barWidth, edgecolor='white', label='2017')
plt.bar(r2, x2018, color='red', width=barWidth, edgecolor='white', label='2018')
plt.bar(r3, x2019, color='green', width=barWidth, edgecolor='white', label='2019')
plt.bar(r4, x2020, color='yellow', width=barWidth, edgecolor='white', label='2020')
plt.bar(r5, x2021, color='black', width=barWidth, edgecolor='white', label='2021')
plt.bar(r6, x2022, color='pink', width=barWidth, edgecolor='white', label='2022')
 
# Add xticks on the middle of the group bars
plt.xlabel('Months', fontweight='bold')
plt.ylabel('# of issues', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(x2017))], labels)

# Create legend & Show graphic
plt.legend()
plt.show()
df2018
```
